```python
import spectral.io.envi as envi
from pathlib import Path
import numpy as np
from spectral import imshow
```

```python
base_path = Path("C:\\Users\\deepe\\Documents\\shubham_projects\\HSI-MSI-Image-Fusion\\hmi_fusion\\artifacts\\brain_image_1")
dark_ref = envi.open(base_path / "darkReference.hdr", base_path / "darkReference")
white_ref = envi.open(base_path / "whiteReference.hdr", base_path / "whiteReference")
data_ref = envi.open(base_path / "raw.hdr", base_path / "raw")
# dark_ref = envi.open(da)
white_nparr = np.array(white_ref.load())
dark_nparr = np.array(dark_ref.load())
data_nparr = np.array(data_ref.load())
```

```python
corrected_nparr = np.divide(
    np.subtract(data_nparr, dark_nparr),
    np.subtract(white_nparr, dark_nparr))
```

```python
imshow(corrected_nparr, (100, 100, 100))
```
